```python
import matplotlib.pyplot as plt

x = range(10)
y1 = [elem*2 for elem in x]
plt.plot(x, y1)

y2 = [elem**2 for elem in x]
plt.plot(x, y2, 'r--')

plt.show()
```

```python
fig_1 = plt.figure(figsize=(8, 4))
ax_1 = fig_1.add_subplot(111)
ax_1.legend(loc=0)
ax_1.set_xlabel('Epoch number')
def train_plot(no):  
    if no == 0:
        color_train, color_valid = 'r', 'grey'
        line_valid = '--'
        x = range(10)
        y1 = [elem*2 for elem in x]
        # Plot the change in the validation and training set accuracy over training.
#         fig_1 = plt.figure(figsize=(8, 4))
#         ax_1 = fig_1.add_subplot(111)
        ax_1.plot(x, y1, label='width 32(train)',color=color_train)
        ax_1.legend(loc=0)
#         ax_1.set_xlabel('Epoch number')
    
        no += 1
    else:
        color_train, color_valid = 'cornflowerblue', 'gold'
        line_valid = '--'     
        x = range(10)
        y2 = [elem**2 for elem in x]
#         fig_1 = plt.figure(figsize=(8, 4))
#         ax_1 = fig_1.add_subplot(111)
        # Plot the change in the validation and training set accuracy over training.
        ax_1.plot(x, y2, label='width 64(train)',color=color_train)
        ax_1.legend(loc=0)

        
for no in range(2):
    train_plot(no)
```

```python
handles,labels = ax_1.get_legend_handles_labels()

handles = [handles[0], handles[1]]
labels = [labels[0], labels[1]]

ax_1.legend(handles,labels,loc=2)
```
